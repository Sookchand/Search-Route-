# Task 2.1 Route Planning
## Problem Descriptions
1.  **State**: Number of cities is 6 ranging from 1 to 6.  Initialization node is number 1.
2.  **Action**: Fromt he initialization node we will move to the next accessible node (1 to 3)
3.  **Current State**: Our final destintion is node 5.  Hence, We would check current node to see if we have reached the destination node(5).
4.  **Path Cost**: We then calculate the total cost from initialization node to the destination node.  
## Implementation and Results

```
!pip install simpleai
!pip install pydot flask
from simpleai.search import SearchProblem, astar, greedy, breadth_first, depth_first, uniform_cost
import math
```

```
COSTS = [
    [0, 7, 9, 'inf', 'inf', 14],
    [7, 0, 10, 15, 'inf', 'inf'],
    [9, 10, 0, 11, 'inf', 2],
    ['inf', 15, 11, 0, 6, 'inf'],
    ['inf', 'inf', 'inf', 6, 0, 9],
    [14, 'inf', 2, 'inf', 9, 0]
]     
```

```
class Route(SearchProblem):

    # initialise with the initial and goal states. Do "-1" because the nodes start from 1.
    def __init__(self, initial, goal):
        self.initial = initial-1
        self.goal = goal-1
        super(Route, self).__init__(initial_state=self.initial)

    # add all connected nodes, i.e. those with a cost of not 0 or inf
    def actions(self, state):
        actions = []
        for action in range(len(COSTS[state])):
            if COSTS[state][action] not in ['inf', 0]:
                actions.append(action)
        return actions

    # the result state is just the action as defined
    def result(self, state, action):
        return action

    # check if goal is reached
    def is_goal(self, state):
        return state == self.goal

    # return the cost between two states
    def cost(self, state, action, state2):
        return COSTS[state][action]
```

## Breadth-First Search (FIFO)

```
problem = Route(1, 5)
result = breadth_first(problem, graph_search=True)

path = [x[1]+1 for x in result.path()] 
print("The route is %s, and total cost is %s" %(path, result.cost))
```

## Depth-First Search (LIFO)

```
problem = Route(1, 5)
result = depth_first(problem, graph_search=True)

path = [x[1]+1 for x in result.path()] 
print("The route is %s, and total cost is %s" %(path, result.cost))
```

## Uniform-Cost Search

```
problem = Route(1, 5)
result = uniform_cost(problem)

# Print the results
path = [x[1]+1 for x in result.path()] 
print("The route is %s, and total cost is %s" %(path, result.cost))
```

## Discussions

In this task, both depth-first search and breadth-first search yielded the same cost of 23 and choose same route[1, 6, 5].  Uninform-cost search used less resource but visited more node.  Also, when all 3 search is compared it can be concluded that uniform-cost search showed in this case to be the "optimal solution."
